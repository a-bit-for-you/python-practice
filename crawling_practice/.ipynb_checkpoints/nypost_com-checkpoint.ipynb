```python
import re
import requests
import time
import random
import json

# user agent is required because some times we may get permission error
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36 Edg/88.0.705.74'
}

# Fetching rss url(xml)
rss = requests.get('https://nypost.com/feed/', headers=headers, timeout=60).text  
rss = rss.split('<item>')

cl_data = []

for i in rss[1:]:
    
# ======= Metacharacters =====
#     \s Returns a match where the string contains a white space character
#     () Capture and group
#     . Any character (except newline character)
#     * Zero or more occurrences
#     ? Zero or one
# =============================

#    flags=re.S is used on multi-line string
#     group method is used to convert object into string
    
    # article url
    source_link = re.search('<link>(.*?)</link>', i, flags=re.S).group(1)
    print(source_link)
    
    source_headline = re.search('<title>(.*?)</title>', i, flags=re.S).group(1)
    print(source_headline)
    
    pub_date_time = re.search('<pubDate>(.*?)</pubDate>', i, flags=re.S).group(1)
    print(pub_date_time)
    
    time.sleep(random.randint(1,3))
    
    # article page(html page)
    page = requests.get(source_link, headers=headers, timeout=60).text
    
    # .*? greedy expression(ignore everything)
    journalist = re.search('<p\s*class="byline">.*?<span>(.*?)</span>', page, flags=re.S).group(1)
    print(journalist)
    
    source_content = re.search('<div\s*class="double-rule">.*?<p>(.*?)<div\s*id="sharebar-trigger-desktop">', page, flags=re.S).group(1)    
    # removing tags
    source_content = re.sub('<.*?>', '', source_content, flags=re.S)
    
    # creating dictionary
    clientdata = {
        'source_link': source_link,
        'source_headline': source_headline,
        'source_content': source_content,
        'journalist': journalist,
        'pub_date_time': pub_date_time
    }
    
    # inserting dictionary to list
    cl_data.append(clientdata)
    
# writing data 
# json.dumps is used to convert list/dict object to string
with open('nypost_com.json', 'w') as f:
    f.write(json.dumps(cl_data, indent=6))
```
