## Linear regression

Linear Regression is a supervised machine learning algorithm. It is used to predict value based on independent variables. It is used to find relationship between variables and forecasting.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20,10)

data =  pd.read_csv('headbrain.csv')
print(data.shape)
data.head()
```

```python
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values

# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
_len = len(X)

# Using the formula to calculate m and c
numer = 0
denom = 0

for i in range(_len):
    numer += (X[i]-mean_x)*(Y[i]-mean_y)
    denom += (X[i]-mean_x) ** 2

# slope
m = numer/denom

# Finding c (y intercept) by using equation of straight line y = mx+c
c = mean_y - (m * mean_x)
m,c
```

```python
max_x = np.max(X)+200
max_y = np.max(Y)+200

x = np.linspace(max_x,max_y)

y = m*x+c

# Regression line
plt.plot(x,y,color='red',label='Regression Line')

# X, Y points
plt.scatter(X,Y,c='blue',label='Scatter Plot')

plt.xlabel('headsize in cm3')
plt.ylabel('Brean Weight in grams')
plt.legend()
plt.show()
```

## Coefficient of Determination-R2 score

### Examples

<b>Case 1 Model gives accurate results</b>

                                        
R2 = 1- 0/200 = 1

<b>Case 2 Model gives ambiguous results</b>

 R2 = 1- 600/200 = -2

```python
# SSres is the sum of squares of the residual errors. (Actual - Predicted)
ss_r = 0

# SStot is the total sum of the errors. (Actual - Mean)
ss_t = 0

for i in range(_len):
    y_pred = m*X[i] + c
    ss_r += (Y[i] - y_pred) ** 2
    ss_t += (Y[i] - mean_y) ** 2
r2 = 1 - (ss_r/ss_t)
print(r2)
```

## Coefficient of Determination-R2 score using scikit-learn

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error

# canot use one-D
X = X.reshape((_len,1))

# creating model
reg = LinearRegression()

# fitting traning data
reg = reg.fit(X,Y)

Y_pred = reg.predict(X)

# calculating R2 Score

r2_score = reg.score(X,Y)
print(r2_score)
```
