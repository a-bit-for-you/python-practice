```python
import pyautogui as pg
import time
from cv2 import *
import sys
import random

start_time = time.time()
time.sleep(1)


# Functions
def press(command):
    time.sleep(random.randint(1,3))
    pg.press(command)
    
    
def hotkey(*command):
    time.sleep(random.randint(1,3))
    pg.hotkey(*command)
    
    
def typewrite(command):
    time.sleep(random.randint(1,3))
    pg.typewrite(command)
    
    
def click(x,y):
    time.sleep(random.randint(1,3))
    pg.click(x,y)
    
    
def click_image(image,confid):
    start_time = time.time()
    while True:
        search = pg.locateOnScreen(image,confidence=confid)
        if search is not None:
            break
        if time.time()-start_time > 50:
            sys.exit(f'{image} not found.')
    cordinates = pg.center(search)
    pg.click(cordinates.x,cordinates.y)
    
    
def wait_for_image(image,confid):
    start_time = time.time()
    while True:
        search = pg.locateOnScreen(image,confidence=confid)
        if search is not None:
            break
        if time.time()-start_time > 50:
            sys.exit(f'{image} not found.')
    cordinates = pg.center(search)
    return True
    

profile_urls = ['https://www.linkedin.com/sales/people/ACwAABlfLj0BboQi7_PP8LROJyBKl7HSu1AKXBk,NAME_SEARCH,DlYH',
                'https://www.linkedin.com/sales/people/ACwAAAJwmLkBmhL_kIVGaawEOSpHJn7c4g6cYo8,NAME_SEARCH,mKG2',
                'https://www.linkedin.com/sales/people/ACwAAABkHVIBSTvu59nod8D-vRy8_8Fix9kMt9c,NAME_SEARCH,-7qn',
                'https://www.linkedin.com/sales/people/ACwAADAa1f8BPt-xQaNTsNxNbx2t3vVkNckUXb8,NAME_SEARCH,L43Z',
                'https://www.linkedin.com/sales/people/ACwAACjnOdMBf2ukjo4Mx3NTug_2p_71fX1EEcw,NAME_SEARCH,qSEr',
                'https://www.linkedin.com/sales/people/ACwAACjj5IQB1zhyvQaQnT3S0nXW7PuibAd4P58,NAME_SEARCH,GZnq']

path = r'F:\python-practice\linked_in_automation\images'
hotkey('ctrl','shift','n')  # opens incognito mode
# login
typewrite('https://www.linkedin.com/sales/login')
press('enter')
click_image(rf'{path}/username.png', 0.7)
typewrite('palaksingh123456.ps@gmail.com')
press('tab')
typewrite('ubsforums@786')
press('enter')
wait_for_image(rf'{path}/logo.png', 0.7)

hotkey('alt', 'd')  # select search bar

for profile_url in profile_urls:
    
    typewrite(profile_url)  # search for profile
    press('enter')
    break
```
